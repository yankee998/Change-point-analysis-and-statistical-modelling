```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Load data

```python
try:
    brent_df = pd.read_csv('C:/Users/Skyline/Change point analysis and statistical modelling/data/raw/BrentOilPrices.csv')
    brent_df['Date'] = pd.to_datetime(brent_df['Date'], format='mixed', dayfirst=True)
    print("Brent Data Preview:")
    print(brent_df.head())
except Exception as e:
    print(f"Error loading BrentOilPrices.csv: {e}")
```

## Load and preprocess events

```python
try:
    events_df = pd.read_csv('C:/Users/Skyline/Change point analysis and statistical modelling/data/processed/events.csv')
    events_df['Date'] = pd.to_datetime(events_df['Date'], format='%Y-%m-%d')
    print("\nEvents Data Preview:")
    print(events_df.head())
except Exception as e:
    print(f"Error loading events.csv: {e}")
```

## Plot price series with events

```python

if 'brent_df' in locals() and 'events_df' in locals():
    plt.figure(figsize=(12, 6))
    plt.plot(brent_df['Date'], brent_df['Price'], label='Brent Oil Price (USD)')
    for _, event in events_df.iterrows():
        plt.axvline(event['Date'], color='red', linestyle='--', alpha=0.5, label=event['Event'] if _ == 0 else "")
    plt.title('Brent Oil Prices with Key Events (1987–2022)')
    plt.xlabel('Date')
    plt.ylabel('Price (USD per barrel)')
    plt.legend()
    try:
        plt.savefig('C:/Users/Skyline/Change point analysis and statistical modelling/notebooks/price_with_events.png')
        print("Plot saved to notebooks/price_with_events.png")
    except Exception as e:
        print(f"Error saving plot: {e}")
    plt.show()
else:
    print("Cannot plot: Missing Brent or events data.")
```
